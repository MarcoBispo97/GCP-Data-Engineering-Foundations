# ☁️ GCP Essentials for Data Scientists (VS Code Terminal)

A curated list of `gcloud`, `gsutil`, and `bq` commands focused on productivity for MLE and Data Science workflows.

---

## 🔑 1. Configuration & Context
* **Initial Login:** `gcloud auth login`
* **Set Current Project:** `gcloud config set project [PROJECT_ID]`
* **Check Current Config (Account/Project):** `gcloud config list`
* **Switch Configurations Quickly:** `gcloud config configurations activate [NAME]`
* **List All Available Projects:** `gcloud projects list`

---

## 📦 2. Cloud Storage (GCS) - `gsutil`
Essential for managing datasets, model artifacts (.pkl, .onnx), and pipeline outputs.

* **List Bucket Contents:** `gsutil ls gs://[BUCKET_NAME]`
* **Upload File:** `gsutil cp [LOCAL_FILE] gs://[BUCKET]`
* **Download Folder (Recursive):** `gsutil cp -r gs://[BUCKET]/data .`
* **Sync Local with Bucket (Deployment/Backup):** `gsutil rsync -r [LOCAL_DIR] gs://[BUCKET]`
* **Check Bucket Size:** `gsutil du -sh gs://[BUCKET]`

---

## 🧠 3. Vertex AI & Machine Learning
Manage training jobs and model deployments without leaving the terminal.

* **List Models in Vertex:** `gcloud ai models list --region=[REGION]`
* **Stream Training Job Logs:** `gcloud ai custom-jobs stream-logs [JOB_ID]`
* **List Notebook Instances:** `gcloud notebooks instances list --location=[REGION]`
* **List Active Endpoints:** `gcloud ai endpoints list --region=[REGION]`

---

## 🏗️ 4. BigQuery - `bq`
High-speed SQL testing and data inspection.

* **Execute Query (Standard SQL):** `bq query --use_legacy_sql=false 'SELECT count(*) FROM `project.dataset.table` LIMIT 10'`
* **List Datasets:** `bq ls`
* **Inspect Table Schema:** `bq show project:dataset.table`
* **Load CSV to BigQuery:** `bq load --source_format=CSV --autodetect dataset.table gs://bucket/data.csv`

---

## 🚢 5. Artifact Registry & Docker (MLOps)
Version control for your Streamlit apps or Model APIs.

* **Authenticate Docker with GCP:** `gcloud auth configure-docker [REGION]-docker.pkg.dev`
* **Tag Image:** `docker tag [LOCAL_NAME] [REGION]-docker.pkg.dev/[PROJECT]/[REPO]/[IMAGE]:[TAG]`
* **Push to Registry:** `docker push [REGION]-docker.pkg.dev/[PROJECT]/[REPO]/[IMAGE]:[TAG]`

---

## 🛠️ 6. Compute Engine (VMs)
* **SSH into Instance (Directly in VS Code):** `gcloud compute ssh [INSTANCE_NAME] --zone=[ZONE]`
* **List Instances (Status/IPs):** `gcloud compute instances list`
* **Start/Stop Instance:** `gcloud compute instances start/stop [INSTANCE_NAME]`

---

## 💡 Productivity Hacks (Pro-Tips)

1. **Project Aliases:** Add this to your `.bashrc` or `.zshrc`:
   `alias gcp-work='gcloud config set project your-project-id-here'`
2. **Clean Output:** Use `--format="table(name, status)"` to filter specific columns.
3. **JSON Filtering:** Pipe to `jq` for automation scripts:
   `gcloud compute instances list --format=json | jq '.[].name'`
