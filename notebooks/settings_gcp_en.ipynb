## Full step-by-step

### 1) Activate `venv` in Git Bash
```bash
cd /c/Users/marco/Documents/code_classes/GCP-Data-Engineering-Foundations
source venv/Scripts/activate
python --version
```

### 2) Temporary PATH fix (current session)
When `sed`, `uname`, and `which` were not found:
```bash
export PATH="/usr/bin:/bin:/mingw64/bin:$PATH"
hash -r
command -v sed
command -v uname
command -v which
```

### 3) Permanent PATH fix in Git Bash
Create `~/.bashrc` with:
```bash
# Stable PATH defaults for Git Bash
export PATH="/usr/bin:/bin:/mingw64/bin:$PATH"

# Google Cloud SDK (user install)
GCLOUD_BIN="/c/Users/marco/AppData/Local/Google/Cloud SDK/google-cloud-sdk/bin"
if [ -d "$GCLOUD_BIN" ]; then
  export PATH="$PATH:$GCLOUD_BIN"
fi
```

Create `~/.bash_profile` with:
```bash
if [ -f "$HOME/.bashrc" ]; then
  . "$HOME/.bashrc"
fi
```

### 4) Find and validate `gcloud` on Windows (PowerShell)
```powershell
Get-Command gcloud -ErrorAction SilentlyContinue | Select-Object -ExpandProperty Source

# Path found in this environment:
# C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd

& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" --version
```

### 5) Sign in to GCP (`gcloud auth login`)
```powershell
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" auth login
```

### 6) Configure ADC for Python libraries
```powershell
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" auth application-default login
```

### 7) Set default project and quota project
```powershell
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" config set project global-grammar-432121-d7
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" auth application-default set-quota-project global-grammar-432121-d7
```

### 8) Validate authentication
```powershell
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" auth list
& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" auth application-default print-access-token
```

> Do not share tokens, OAuth verification codes, or private keys.

## Issues found and root cause

1. **`sed` / `uname` / `which` returning `command not found`**
   - **Cause:** inconsistent `PATH` in Git Bash during/after environment activation.
   - **Fix:** restore `/usr/bin:/bin:/mingw64/bin` and persist it in `~/.bashrc`.

2. **`invalid_grant: Invalid code verifier` in no-browser login**
   - **Cause:** reusing an OAuth code from a previous session/link.
   - **Fix:** use the fresh flow from the current session (or run `gcloud auth login` with browser).

## Security best practices

- Never share OAuth verification codes.
- Never share access tokens / refresh tokens.
- Never commit service account JSON keys to GitHub.
- Prefer ADC for local development whenever possible.
- Apply least-privilege IAM permissions.
