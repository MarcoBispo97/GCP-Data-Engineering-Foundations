### 1) Ativar o `venv` no Git Bash
```bash
cd /c/Users/marco/Documents/code_classes/GCP-Data-Engineering-Foundations
source venv/Scripts/activate
python --version
```

### 2) Corrigir PATH temporariamente (sessão atual)
Quando `sed`, `uname` e `which` não eram encontrados:
```bash
export PATH="/usr/bin:/bin:/mingw64/bin:$PATH"
hash -r
command -v sed
command -v uname
command -v which
```

### 3) Corrigir PATH permanentemente no Git Bash
Criar `~/.bashrc` com:
```bash
# Stable PATH defaults for Git Bash
export PATH="/usr/bin:/bin:/mingw64/bin:$PATH"

# Google Cloud SDK (user install)
GCLOUD_BIN="/c/Users/marco/AppData/Local/Google/Cloud SDK/google-cloud-sdk/bin"
if [ -d "$GCLOUD_BIN" ]; then
  export PATH="$PATH:$GCLOUD_BIN"
fi
```

Criar `~/.bash_profile` com:
```bash
if [ -f "$HOME/.bashrc" ]; then
  . "$HOME/.bashrc"
fi
```

### 4) Encontrar e validar `gcloud` no Windows (PowerShell)
```powershell
Get-Command gcloud -ErrorAction SilentlyContinue | Select-Object -ExpandProperty Source

# Caminho encontrado no ambiente:
# C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd

& "C:\Users\marco\AppData\Local\Google\Cloud SDK\google-cloud-sdk\bin\gcloud.cmd" --version
```

### 8) Validar autenticação

## Problemas encontrados e causa raiz

1. **`sed` / `uname` / `which` como `command not found`**
   - **Causa:** `PATH` inconsistente no Git Bash durante/após ativação do ambiente.
   - **Correção:** restaurar `/usr/bin:/bin:/mingw64/bin` e persistir em `~/.bashrc`.

2. **`invalid_grant: Invalid code verifier` no login sem navegador**
   - **Causa:** reutilização de código OAuth de uma sessão/link anterior.
   - **Correção:** usar o fluxo novo da sessão atual (ou `gcloud auth login` com navegador).

## Boas práticas de segurança

- Nunca compartilhar código de verificação OAuth.
- Nunca compartilhar access token / refresh token.
- Nunca versionar chave JSON de service account no GitHub.
- Preferir ADC para desenvolvimento local quando possível.
- Limitar permissões IAM ao mínimo necessário.
